```python
#Import dependencies
import numpy as np
import pandas as pd
from citipy import citipy
import requests

#Import the API key.
from config import weather_api_key
```

```python
#Create a set of random latitude and longitude combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=2000)
longs = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lats, longs)
coordinates = list(lat_lngs)
len(coordinates)
```

```python
#Get the nearest city for each latitude and longitude combination
cities = []
for coordinate in coordinates:
    
    city = citipy.nearest_city(coordinate[0], coordinate[1]).city_name

    #If the city is unique, add it to the cities list
    if city not in cities:
        cities.append(city)

len(cities)
```

```python
#Perform API call in OpenWeatherMap to retrieve weather data
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 
city_data = []
record_count = 1
set_count = 1

#Beginning message print
print("Beginning Data Retrieval\n-------------------------")

for i, city in enumerate(cities):
    
    #Group into sets of 50
    if (i % 50 == 0 and i >= 50):
        set_count += 1
    
    #Create endpoint URL with each city, removing blank spaces - concatenate names to reduce liklihood of name errors
    city_url = url + "&q=" + city.replace(" ","+")
    
    #Log the URL, record, set numbers and the city
    print(f"Processing Record {record_count} of Set {set_count} | {city}")
    record_count += 1
    
    #Run API requests for each of the cities
    try:
        #Parse the JSON and retrieve data
        city_weather = requests.get(city_url).json()
        
        #Parse out the needed data
        city_lat = city_weather["coord"]["lat"]
        city_lng = city_weather["coord"]["lon"]
        city_max_temp = city_weather["main"]["temp_max"]
        city_humidity = city_weather["main"]["humidity"]
        city_clouds = city_weather["clouds"]["all"]
        city_wind = city_weather["wind"]["speed"]
        city_country = city_weather["sys"]["country"]
        city_description = city_weather["weather"][0]["description"]
        
        #Append the city information into city_data dictionary
        city_data.append({"City": city.title(),
                          "Country": city_country,
                          "Lat": city_lat,
                          "Lng": city_lng,
                          "Max Temp": city_max_temp,
                          "Humidity": city_humidity,
                          "Cloudiness": city_clouds,
                          "Wind Speed": city_wind,
                          "Description": city_description})

    #If an error occurs, skip the city and keep going
    except:
        print(f"City name {city} not found. Skipping...")
        pass

# Indicate that Data Loading is complete.
print("------------------------\nData Retrieval Complete\n------------------------")
```

```python
# Check count of available cities
len(city_data)
```

```python
#Convert city weather data to DataFrame
city_weather_df = pd.DataFrame(city_data)
city_weather_df.head()
```

```python
#Export weather data to CSV file "WeatherPy_Database"
output_data_file = "WeatherPy_Database.csv"
city_weather_df.to_csv(output_data_file, index_label="City_ID")
```
